```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gini_coefficients = pd.read_csv(f'../data/estat_tessi190_en.csv')
poverty = pd.read_csv(f'../data/estat_sdg_01_10_en.csv')
```

```python
poverty[poverty['Geopolitical entity (reporting)'] == "Czechia"].head()
```

```python
# Example: Gini over time by country
plt.figure(figsize=(12,6))
sns.lineplot(data=gini_coefficients[gini_coefficients['Geopolitical entity (reporting)'] == "Czechia"], x='TIME_PERIOD', y='OBS_VALUE', hue='Geopolitical entity (reporting)')
plt.title('Gini Coefficient Over Time')
plt.ylabel('Gini Coefficient')
plt.show()
```

```python
# Example: Gini over time by country
plt.figure(figsize=(12,6))
sns.lineplot(data=poverty[poverty['Geopolitical entity (reporting)'] == "Czechia"], x='TIME_PERIOD', y='OBS_VALUE', hue='Geopolitical entity (reporting)')
plt.title('Gini Coefficient Over Time')
plt.ylabel('Gini Coefficient')
plt.show()
```
